```python
import os 
import numpy as np
import pandas as pd 
import sklearn.metrics as metrics
```

```python
folder = os.path.join(
    os.environ["SYNC"], r"MotionData\data\2nd-Development", "parsedData"
)

event_name = {
    "0-SLW": 1,
    "1-MLW": 2,
    "2-FLW": 3,
    "3-RD": 4,
    "4-SD": 5,
    "5-sit": 6,
    "6-stand": 7,
    "7-RA": 8,
    "8-SA": 9,
}

event_table = [(a, b) for a, b in event_name.items()]
event_table
```

```python
def summary_cr(fname):
    with open(os.path.join(folder, fname), "rb") as f:
        y = np.load(f)
        p = np.load(f)

    cr = metrics.classification_report(y_true=y, y_pred=p, output_dict=True)
    df = pd.DataFrame(cr).transpose()
    
#     table_dom = df.to_html(f'{fname}.html', classes='pure-table')
    table_dom = df.to_html(classes='pure-table')
    
    html = open('template.html').read()
    html = html.replace('---- Table goes here ----', table_dom)
    with open(f'{fname}.html', 'w') as f:
        f.write(html)
        
    return df
```

```python
fname = 'raw-svc.npy'
df = summary_cr(fname)
df
```
